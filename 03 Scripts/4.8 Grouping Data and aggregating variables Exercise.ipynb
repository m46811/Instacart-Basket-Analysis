#### Grouping Data and Aggregating new variables

```python
# Import libraries & Declare the path to the data files
import pandas as pd
import numpy as np
import os
path_data = r'C:\Users\PC\Desktop\CareerFoundry Data Analytics\08-2025 Instacart Basket Analysis'
```

```python
# Import data 
df = pd.read_pickle(os.path.join(path_data, '02 Data', 'Prepared Data','ords_prods_merged_busiest_days_and_hours.pkl'))
```

```python
# Create a subset (slice) to avoid memory issues
df = df[:1000000]
```

```python
df.groupby('product_name')
```

```python
# df.groupby('department_id').agg({'order_number': ['mean']})
df.groupby('department_id')['order_number'].mean()
```

```python
df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
df.head(15)
```

```python
# This command tells pandas not to assign any options regarding the maximum number of rows to display. 
pd.options.display.max_rows = None
```

```python
# Assign a “loyalty” label to a user ID based on its corresponding max order value.
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <= 40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```
