#### Intro to Visualization with Python Part 2

```python
## Import libraries 
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy

path_data = r'C:\Users\PC\Desktop\CareerFoundry Data Analytics\08-2025 Instacart Basket Analysis'
```

```python
## Import data 
df = pd.read_pickle(os.path.join(path_data, '02 Data', 'Prepared Data', 'ords_prods_customers.pkl'))
```

# Visualisations

## Histogram 

```python
# Create a histogram for'order_hour_of_day'
hist = df['order_hour_of_day'].plot.hist(bins = 24)
```

The busiest time of the day is between 8:00 and 18:00.

## Bar chart

```python
# Create a bar chart for "loyalty_flag"
bar = df['loyalty_flag'].value_counts().plot.bar()
```

## Line Charts

```python
# Create a sample of the dataframe for the line charts
np.random.seed(4)
dev = np.random.rand(len(df)) <= 0.7
big = df[dev]
small = df[~dev]
```

```python
df_linechart = small 
```

```python
# Group by hour of the day and calculate the mean expenditure
hourly_expenditure = df_linechart.groupby('order_hour_of_day')['prices'].mean().reset_index()
```

```python
# Create a line chart using the 'order_hour_of_day' and 'prices' columns
line_hour_price = sns.lineplot(data = df_linechart, x = 'order_hour_of_day', y = 'prices')
```

```python
hourly_expenditure.head(24)
```

There are no major fluctuations, except for a peak at 2:00. Further checks are necessary to identify potential outliers.

```python
# Create another line chart for the age and the number of dependets 
line_age_dep = sns.lineplot(data = df_linechart, x = 'Age', y = 'Number of Dependents')
```

The y-axis ranges between 1.44 and 1.58 dependents, which suggests that on average, people across different ages have about 1.5 dependents. The chart is highly fluctuating with sharp ups and downs, meaning there isn’t a smooth or clear trend between age and the number of dependents.

## Scatterplot

```python
# Create a scatterplot for 'Age' and 'Income' using the data sample
scatter = sns.scatterplot(x = 'Age', y = 'Income', data = df_linechart)
```

This scatterplot shows that most people earn below 200K across all ages, but older individuals (above 40) tend to have more variation.
