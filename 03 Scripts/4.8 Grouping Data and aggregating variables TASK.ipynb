#### Grouping Data and Aggregating new variables

```python
## Import libraries
import pandas as pd
import numpy as np
import os
path_data = r'C:\Users\PC\Desktop\CareerFoundry Data Analytics\08-2025 Instacart Basket Analysis'
```

```python
## Import data 
df_ords_prods_merge = pd.read_pickle(os.path.join(path_data, '02 Data', 'Prepared Data','ords_prods_merged_busiest_days_and_hours.pkl'))
```

```python
df_ords_prods_merge.shape
```

## Aggregated Mean Calculation for the Dataframe

```python
# Group by 'department_id' and 'order_number' and calculate Mean using agg() function
df_ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

## How do the results for the entire dataframe differ from those of the subset?
The full dataframe shows higher average order numbers across most departments, suggesting it reflects a wider range of customer behavior compared to the subset. Overall, the full dataframe indicates more frequent ordering behavior across departments than the subset.

## Create a Loyalty Flag

```python
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check the Output
df_ords_prods_merge.head(20)
```

```python
# Creating a 'Loyalty' Label Flag using loc Function
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check Value Counts on 'loyalty_flag'
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Check the Output
df_ords_prods_merge.head(20)
```

## Spending Habits by Loyalty Category

```python
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

New customers tend to spend the most on average. This could suggest that they are more willing to try out higher-priced products or there might be introductory offers that encourage higher spending. Regular customers spend slightly less than new customers (about 0.8 less), but more than loyal customers. This group sits in the middle. Loyal customers actually spend the least on average. 

## Create a Spending Flag

```python
# Aggregate Data with transform() Function
df_ords_prods_merge['avg_product_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Check the relevant columns 
df_ords_prods_merge[['user_id', 'avg_product_price']].head(20)
```

```python
# Create a 'Spending' Label Flag using the loc function
df_ords_prods_merge.loc[df_ords_prods_merge['avg_product_price'] >= 10, 'spending_flag'] = 'High spender'
df_ords_prods_merge.loc[df_ords_prods_merge['avg_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Check value count on 'spending_flag' column
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Check Releavent Columns
df_ords_prods_merge[['user_id', 'avg_product_price', 'spending_flag']].head(20)
```

## Create an Order Frequency Flag

```python
# Aggregate the Data with transform() Function
df_ords_prods_merge['median_days_between_orders'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Check the relevant columns 
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(20)
```

```python
# Create a 'Order Frequency' Label flag using loc function
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_between_orders'] <= 20) & (df_ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency_flag'] = 'Regular customer'
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts on the 'order_frequency_flag'
df_ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```

```python
# Check the relevant columns again
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency_flag']].head(20)
```
