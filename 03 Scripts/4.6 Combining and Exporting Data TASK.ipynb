```python
# Import libraries & Declare the path to the data files
import pandas as pd
import numpy as np
import os
path = r'C:\Users\PC\Desktop\CareerFoundry Data Analytics\08-2025 Instacart Basket Analysis\02 Data\Original Data'
path_data = r'C:\Users\PC\Desktop\CareerFoundry Data Analytics\08-2025 Instacart Basket Analysis'
```

```python
# Import data
df_orders_products_combined = pd.read_pickle(os.path.join(path_data, '02 Data', 'Prepared Data','orders_products_combined.pkl'))
df_prod = pd.read_csv(os.path.join(path_data,'02 Data', 'Prepared Data','products_cleaned.csv'), index_col=False)
```

```python
# Test merge without overwriting with merge flag 'merge_new'
pd.merge(df_orders_products_combined, df_prod, on = ['product_id'] , how = 'inner', indicator = 'merge_new' )
```

```python
# Remove the '_merge' column, as it is no longer needed and another indicatior is in use. 
df_orders_products_combined.drop('_merge', axis = 1 )
```

```python
ords_prods_merged = df_orders_products_combined.merge(df_prod, on = ['product_id'] , indicator = 'merged')
```
